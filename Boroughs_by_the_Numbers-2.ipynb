```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
missing_values = ["s"]
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values=missing_values)
```

City-wide Pass Rate by Exam 

```python
#Regents passing rates for all reporting schools
```

```python
df.groupby(['Regents Exam'])['Percent Scoring 65 or Above'].mean().plot(kind='bar')
```

Pass Rate by Borough: Bronx

```python
#Passing rates for Bronx schools fell below passing rates for New York City in all exams.
```

```python
df[df['School DBN'].str.contains('X')].groupby(['Regents Exam'])['Percent Scoring 65 or Above'].mean().plot.bar()
```

Pass Rate by Borough: Brooklyn

```python
#Brooklyn schools kept pace with citywide scores, with differences in percentages faliing within no more than five points.
```
